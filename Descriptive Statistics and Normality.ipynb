```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#1:  
normal = np.random.normal(0, 1, 100)
bernoulli= np.random.binomial(1, .5, 100)
binomial = np.random.binomial(20, 0.5, 100)
gamma = np.random.gamma(5,1, 100)
poisson = np.random.poisson(3, 100)
logistic = np.random.logistic(10, 1, 100)

#2,#3:
plt.hist(normal)
plt.axvline(normal.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean() - normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

plt.hist(bernoulli)
plt.axvline(bernoulli.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(bernoulli.mean() + normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(bernoulli.mean() - normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

plt.hist(binomial)
plt.axvline(normal.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normal.mean() + normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normal.mean() - normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

plt.hist(gamma)
plt.axvline(binomial.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(binomial.mean() + normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(binomial.mean() - normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

plt.hist(poisson)
plt.axvline(poisson.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(poisson.mean() + normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(poisson.mean() - normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

plt.hist(logistic)
plt.axvline(logistic.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(logistic.mean() + normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(logistic.mean() - normal.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()

#3

```

The descriptive statistics of mean and standard deviation are only useful for the normal distribution.  In the others, you can't determine anything useful about probabilities

```python
normal1 = np.random.normal(5, .5, 100)
normal2 = np.random.normal(10, 1, 100)
normal3 = normal1 + normal2



plt.hist(normal3)
plt.axvline(normal3.mean(), color='r', linestyle='solid', linewidth=2)
plt.axvline(normal3.mean() + normal3.std(), color='r', linestyle='dashed', linewidth=2)
plt.axvline(normal3.mean() - normal3.std(), color='r', linestyle='dashed', linewidth=2)
plt.show()
```

```python
var1 = np.random.normal(5,.5, 100)
var2 = np.random.normal(10,1, 100)
var3 = var1+var2

mean = np.mean(var3)
sd = np.std(var3)

plt.hist(var3)
plt.axvline(x=mean,color='black')
plt.axvline(x=mean+sd, color='red')
plt.axvline(x=mean-sd, color='red')
plt.show()
print('Results will vary depending on random seed.  In general, stats should show some skew.')
```
